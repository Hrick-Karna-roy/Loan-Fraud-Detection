### Import the Libreries ###

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import pickle
```

## importing the dataset

```python
dataset = pd.read_csv('/home/hrick/Desktop/Training Data.csv')

dataset.head()
```

```python
dataset.drop(['Id'],axis  = 1,inplace  = True)
```

## Dropping the city feature for now

```python
dataset.drop(['CITY'],axis  = 1,inplace  = True)
```

```python
dataset.head()
```

## Feature Encoding

```python
dataset["Married/Single"].replace('single', 0, inplace=True)
dataset["Married/Single"].replace('married', 1, inplace=True)
dataset["Married/Single"] = dataset["Married/Single"].astype(int)

dataset.head()
```

```python
dataset["House_Ownership"].replace('rented', 0, inplace=True)
dataset["House_Ownership"].replace('norent_noown', -1, inplace=True)
dataset["House_Ownership"].replace('owned', 1, inplace=True)
dataset["House_Ownership"] = dataset["House_Ownership"].astype(int)

dataset.head()
```

```python
dataset["Car_Ownership"].replace('yes', 1, inplace=True)
dataset["Car_Ownership"].replace('no', 0, inplace=True)
dataset["Car_Ownership"] = dataset["Car_Ownership"].astype(int)

dataset.head()
```

## label encoding for instance

```python
labelEncoder = LabelEncoder()
```

```python
data = dataset
# print(data.head())
# Encode the object data to type int
for e in data.columns:
    if data[e].dtype == 'object':
        labelEncoder.fit(list(data[e].values))
        data[e] = labelEncoder.transform(data[e].values)
        
        # Accommodate the data that has been changed
        tData = data
```

```python
tData.head()
```

```python
tData.corr()
```

## one hot encoding the State feature

```python
intrdata = pd.get_dummies(dataset[['Profession','STATE']])
```

### visualizing the one hot encoded state feature data

```python
intrdata.head()
```

```python
dataset = pd.concat([dataset,intrdata],axis = 1)
```

## train test split of dataset

```python
label = tData['Risk_Flag']
```

```python
tData.drop(['Risk_Flag'],axis = 1,inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(tData, label, test_size=0.33, random_state=42)
```

## logistic regression 

```python
clf = LogisticRegression().fit(X_train, y_train)
```

```python
clf.predict(X_train[:10])
```

```python
clf.predict_proba(X_test)
```

## score on the test data

```python
clf.score(X_test, y_test)
```
